```python
data=np.load('data/qiskit_circuit_energy_estimation_results.npz')
print(data)
energies=data['energies']
errors=data['errors']
hamiltonian_terms=data['hamiltonian_terms']
```

```python
plt.plot(energies[:-1])
plt.scatter(np.arange(len(energies[:-1])),energies[:-1],label='QPU')
plt.scatter(np.arange(len(energies_exact_dt_02[:-1])),energies_exact_dt_02[:-1],label=r'CPU $\Delta t \omega=0.2$ ')
plt.plot(energies_exact_dt_02[:-1])


plt.scatter(np.arange(len(energies[:-1])),energies_exact[:-1],label='CPU')
plt.plot(energies_exact[:-1])

plt.xticks(np.arange(len(energies[:-1])),labels_exact[:-1])
plt.legend(fontsize=20)
plt.ylabel('Energy (MeV)',fontsize=20)
plt.show()
```
